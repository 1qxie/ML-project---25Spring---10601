```python
donations_df = pd.read_csv("/Users/santiago9107/Desktop/ML Project/donations.csv")
essays_df = pd.read_csv("/Users/santiago9107/Desktop/ML Project/essays.csv")
outcomes_df = pd.read_csv("/Users/santiago9107/Desktop/ML Project/outcomes.csv")
projects_df = pd.read_csv("/Users/santiago9107/Desktop/ML Project/projects.csv")
resources_df = pd.read_csv("/Users/santiago9107/Desktop/ML Project/resources.csv")
sampleSubmission_df = pd.read_csv("/Users/santiago9107/Desktop/ML Project/sampleSubmission.csv")
```

```python
filtered_df = donations_df[
    (donations_df['donation_to_project'] >= 0) & 
    (donations_df['donation_to_project'] <= 500)
]

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(filtered_df['donation_to_project'], bins=50, kde=True)
plt.title('Distribution donation_to_project (0-500)')

plt.subplot(1, 2, 2)
sns.boxplot(x=filtered_df['donation_to_project'])
plt.title('donation_to_project Boxplot (0-500)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=matrix_df, x="avg_unit_price", hue="label", bins=200, kde=True, palette="Set1", stat="count")
plt.title("Distribution of avg_unit_price by Funding Outcome")
plt.xlabel("Average Unit Price")
plt.ylabel("Count")
plt.xlim(0, 1200)  # Adjust if needed to reduce outlier impact
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(
    data=matrix_df,
    x="students_reached",
    hue="label",
    bins=100,
    kde=True,
    stat="count",
    palette="Set1"
)
plt.title("Distribution of students_reached by Funding Outcome")
plt.xlabel("Students Reached")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(
    data=matrix_df,
    x="total_quantity",
    hue="label",
    bins=100,
    kde=True,
    stat="count",
    palette="Set1"
)
plt.title("Distribution of total_quantity by Funding Outcome")
plt.xlabel("Total Quantity")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(
    data=matrix_df,
    x="unique_items",
    hue="label",
    bins=100,
    kde=True,
    stat="count",
    palette="Set1"
)
plt.title("Distribution of unique_items by Funding Outcome")
plt.xlabel("Unique Items")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(
    data=matrix_df,
    x="total_price_excluding_optional_support",
    hue="label",
    bins=100,
    kde=True,
    stat="count",
    palette="Set1"
)
plt.title("Distribution of total_price_excluding_optional_support by Funding Outcome")
plt.xlabel("Total Price (Excl. Optional Support)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(matrix_df.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.boxplot(data=matrix_df, x="label", y="total_price_excluding_optional_support")
plt.title("Total Price vs Funding Outcome")
plt.show()
```
